Cumulative Daily Returns

```python
Bitcoin['Cumulative Return'] = (1 + Bitcoin['returns']).cumprod()
```

```python
Ethereum['Cumulative Return'] = (1 + Ethereum['returns']).cumprod()
Ethereum_Classic['Cumulative Return'] = (1 + Ethereum_Classic['returns']).cumprod()
Wrapped_Bitcoin['Cumulative Return'] = (1 + Wrapped_Bitcoin['returns']).cumprod()
Litecoin['Cumulative Return'] = (1 + Litecoin['returns']).cumprod()
```

```python
Bitcoin['Cumulative Return'].plot(label='bitcoin',figsize=(16,8),title='Cumulative Return')
Ethereum['Cumulative Return'].plot(label='ethereum')
Ethereum_Classic['Cumulative Return'].plot(label='ethereum_Classic')
Litecoin['Cumulative Return'].plot(label='litecoin')
Wrapped_Bitcoin['Cumulative Return'].plot(label='wrapped_Bitcoin')
plt.legend()
```
